```python
import pandas as pd
import matplotlib.pyplot as plt
X = pd.DataFrame(iris.data, columns=['sepal length','sepal width', 'petal length','petal width'])
y = pd.DataFrame(iris.target, columns=['class'])
```

2 Consider only two dimensions and visualize the classes

```python
# print(iris.data[:10])
# print(iris.data[:10,0])
colors = ['teal', 'yellowgreen', 'gold']
for i in range(len(iris.target)):
    if iris.target[i]==0:
#         print('0:',i)
        plt.scatter(iris.data[i,0],iris.data[i,2],color=colors[0])
    elif iris.target[i]==1:
#         print('1:',i)
        plt.scatter(iris.data[i,0],iris.data[i,2],color=colors[1])
    elif iris.target[i]==2:
#         print('2:',i)
        plt.scatter(iris.data[i,0],iris.data[i,2],color=colors[2])
#print(iris.target)
```
